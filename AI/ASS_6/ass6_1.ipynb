```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("E:\SUMMER SEM\AI\LABS\ASS_6\data.csv")
df.info()
```

```python
df.experience.unique()
```

```python
df.experience.replace({'one':1, 'two':2, 'three': 3, 'four':4, 'five': 5, 'six':6, 'seven':7, 'eight':8,'nine':9,'ten':10, 'eleven':11},inplace=True)
```

```python
mean_value=df['written_score'].mean()
df['written_score'].fillna(value=mean_value, inplace=True)

mean_value=df['interview_score'].mean()
df['interview_score'].fillna(value=mean_value, inplace=True)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))
plt.scatter(df['experience'],df['written_score'],c=df['salary'],marker='*',s=1000,edgecolors='red')
for i, salary in enumerate (df['salary']):
    plt.text(df.experience[i]+0.1, df['written_score'][i]+0.1, salary)

plt.title(label='Experience v/s Test_Score')
plt.xlabel(xlabel='Experience(years)')
plt.ylabel(ylabel='written_score')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))
plt.scatter(df['experience'],df['interview_score'],c=df['salary'],marker='*',s=1000,edgecolors='red',)
for i, salary in enumerate (df['salary']):
    plt.text(df.experience[i]+0.1, df['interview_score'][i]+0.1, salary)

plt.title(label='Experience v/s Interview_Score')
plt.xlabel(xlabel='Experince(years)')
plt.ylabel(ylabel='interview_score')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))
plt.scatter(df['written_score'],df['interview_score'],c=df['salary'],marker='*',s=1000,edgecolors='red')
for i, salary in enumerate (df['salary']):
    plt.text(df['written_score'][i]+0.1, df['interview_score'][i]+0.1, salary)

plt.title(label='Test_Score v/s Interview_score')
plt.xlabel(xlabel='written_score')
plt.ylabel(ylabel='Interview_Score')
plt.show()
```

```python
x = df.loc[:,df.columns!='salary']         
y = df['salary']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=3)
```

```python
knn3 = KNeighborsClassifier(n_neighbors = 4)
knn2 = KNeighborsClassifier(n_neighbors=1)
```

```python
knn3.fit(x_train, y_train)
knn2.fit(x_train, y_train)

y_pred_3 = knn3.predict(x_test)
y_pred_2 = knn2.predict(x_test)
```

```python
print("Evaluation report for KNN with 3 neighours:")
print(accuracy_score(y_test, y_pred_3))
print(classification_report(y_test, y_pred_3))
```

```python
print("Evaluation report for KNN with 2 neighours:")
print(accuracy_score(y_test, y_pred_2))
print(classification_report(y_test, y_pred_2))
```
